```python
import numpy as np
import matplotlib.pyplot as plt

def ejecutar_jacobi(matriz, constantes, inicio, limite=100, tolerancia=1e-5):
    mat = np.array(matriz, dtype=float)
    vec = np.array(constantes, dtype=float)
    actual = np.array(inicio, dtype=float)
    
    dim = len(mat)
    camino = [actual.copy()]
    
    for _ in range(limite):
        siguiente = np.zeros_like(actual)
        
        for f in range(dim):
            sigma = 0
            for c in range(dim):
                if f != c:
                    sigma += mat[f, c] * actual[c]
            
            siguiente[f] = (vec[f] - sigma) / mat[f, f]
            
        camino.append(siguiente.copy())
        
        if np.linalg.norm(siguiente - actual) < tolerancia:
            return np.array(camino)
            
        actual = siguiente.copy()
        
    return np.array(camino)
```

### Ejercicio 1

```python
mat_a = np.array([[2, 1], [3, 2]])
vec_a = np.array([5, 8])
ini_a = np.array([0, 0])

res_1 = ejecutar_jacobi(mat_a, vec_a, ini_a)

print("Resultados Caso 1:")
print(res_1)
print(f"Converge en: {res_1[-1]}")
```

### Ejercicio 2

```python
mat_b = np.array([[1, 1], [-2, 1]])
vec_b = np.array([6, 0])
ini_b = np.array([0, 0])

res_2 = ejecutar_jacobi(mat_b, vec_b, ini_b)

print("Resultados Caso 2:")
print(res_2)
print(f"Converge en: {res_2[-1]}")
```
